```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score 
```

```python
data=pd.read_csv('insurance.csv')
data.head()
```

```python
data.info()
```

```python
data.head(20)
```

```python
data['sex'].replace('female', 0, inplace=True)
data['sex'].replace('male', 1, inplace=True)
data['sex']=data['sex'].astype('int64')
```

```python
data['smoker'].replace('no', 0, inplace=True)
data['smoker'].replace('yes', 1, inplace=True)
data['smoker']=data['smoker'].astype('int64')
```

```python
region = pd.get_dummies(data['region'], prefix='region', dtype=int)
region.head()
```

```python
X = pd.concat([data[['age', 'sex', 'bmi', 'children', 'smoker']], region], axis = 1) 
X.head()
```

```python
y=data['charges']
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=7)
```

```python
def mse(X,y,vec):
    p=len(y)
    ms=(1./p)*np.sum((X.dot(vec)-y)**2)
    return ms
def grad(X,y,learn,i):
    X = np.hstack((np.ones((X.shape[0], 1)), X)) 
    params = np.random.rand(X.shape[1])
    m = X.shape[0]
    
    cost_track = np.zeros((i, 1))
    
    for i in range(i):
        params = params - (2./m * learn * np.dot(X.T, (np.dot(X, params) - y)))
        cost_track[i] = mse(X, y, params)
    
    return cost_track, params
    
```

```python
def predict(X,params):
    X=np.hstack((np.ones((X.shape[0], 1)), X))
    np.dot(X,params)
    return np.dot(X,params)
```

```python
teack, weights = grad(X_train, y_train, 0.0003, 100000)

print(weights)

y_pred = predict(X_test, weights)
r2 = r2_score(y_test, y_pred)

print("R-squared score:", r2)
```
