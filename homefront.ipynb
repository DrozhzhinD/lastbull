```python
import random
```

```python
class Weapon:
    def __init__(self, damage, durability):
        self.damage = damage
        self.durability = durability

    def use_weapon(self):
        if self.durability > 0:
            self.durability -= 1
            return self.damage
        else:
            return 0

class Sword(Weapon):
    def __init__(self):
        super().__init__(damage=10, durability=40)

class Bow(Weapon):
    def __init__(self):
        super().__init__(damage=15, durability=30)

class Crossbow(Weapon):
    def __init__(self):
        super().__init__(damage=20, durability=50)
```

```python
class Soldier:
    def __init__(self, name, allegiance):
        self.name = name
        self.allegiance = allegiance
        self.health = 100
        self.max_damage = 25
        self.weapon = random.choice([Sword(), Bow(), Crossbow()])

    def attack_damage(self):
        return random.randint(0, self.max_damage) + self.weapon.use_weapon()
```

```python
class Army:
    def __init__(self, allegiance, soldier_count):
        self.allegiance = allegiance
        self.soldiers = [Soldier(f'{allegiance}_{i}', allegiance) for i in range(soldier_count)]

    def attack_army(self, enemy_army):
        total_damage = 0
        for soldier in self.soldiers[:]:
            if enemy_army.soldiers:
                target = random.choice(enemy_army.soldiers)
                damage = soldier.attack_damage()
                target.health -= damage
                total_damage += damage
                print(f"{self.allegiance} солдат \"{soldier.name}\" атаковал {enemy_army.allegiance} солдат \"{target.name}\" нанёс {damage} урона {enemy_army.allegiance} члену армии \"{target.name}\"  {target.health} здоровья осталось")
                if target.health <= 0:
                    print(f"{target.name} из {enemy_army.allegiance} армия погиб!")
                    enemy_army.soldiers.remove(target)
            else:
                break
        return total_damage
```

```python
def epic_battle(*armies):
    armies = list(armies) 
    while len(set(army.allegiance for army in armies if army.soldiers)) > 1:
        random.shuffle(armies)
        for i in range(len(armies)):
            if i+1 < len(armies) and armies[i].soldiers and armies[i+1].soldiers and armies[i].allegiance != armies[i+1].allegiance:
                armies[i].attack_army(armies[i+1])
    surviving_armies = [army for army in armies if army.soldiers]
    
    winning_allegiance = surviving_armies[0].allegiance
    return winning_allegiance

army_red = Army('красный', 20)
army_blue = Army('синий', 20)

victorious_allegiance = epic_battle(army_red, army_blue)
print(f"Победившая армия: {victorious_allegiance}")
```
