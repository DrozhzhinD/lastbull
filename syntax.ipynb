```python
import spacy
```

```python
with open('na-dne1902.txt', 'r', encoding='utf-8', errors='ignore') as file: 
    text1 = file.read()
with open('chelkash1904.txt', 'r', errors='ignore') as file: 
    text2 = file.read()  
with open('deti1905.txt', 'r', encoding='utf-8', errors='ignore') as file: 
    text3 = file.read() 
with open('mat1906.txt', 'r', errors='ignore') as file: 
    text4 = file.read() 
with open('detstvo1913.txt', 'r', errors='ignore') as file: 
    text5 = file.read() 
with open('univer1922.txt', 'r', errors='ignore') as file: 
    text6 = file.read() 
```

```python
nlp = spacy.load('ru_core_news_sm')
doc = nlp(text1)

sentences = list(doc.sents)[:100]
for sentence in sentences:
    for token in sentence:
        print(token.text, token.pos_, token.dep_)
```

```python
counter = 0

for sentence in sentences:
    for token in sentence:
        if token.dep_ == 'ROOT':
            print(token.text)
            counter += 1
            if counter >= 100:
                break
    if counter >= 100:
        break

def count_root_speech_parts():
        res = dict()
        for token in doc:
            if token.pos_ not in ("SPACE", "PUNCT") and token.dep_ == "ROOT": 
                if token.pos_ not in res.keys():
                    res[token.pos_] = 1
                else:
                    res[token.pos_] += 1
        root_count = res
        root_count_sorted = dict(sorted(root_count.items(), key=lambda item: item[1]))
        return root_count_sorted
count_root_speech_parts()
```
