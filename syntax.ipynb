```python
import spacy
```

```python
with open('univer1922.txt', 'r', errors='ignore') as file: 
    text5 = file.read()
```

```python
nlp = spacy.load('ru_core_news_sm')
doc = nlp(text5)

sentences = list(doc.sents)[:100]
for sentence in sentences:
    for token in sentence:
        print(token.text, token.pos_, token.dep_)
```

```python
counter = 0

for sentence in sentences:
    for token in sentence:
        if token.dep_ == 'ROOT':
            print(token.text)
            counter += 1
            if counter >= 100:
                break
    if counter >= 100:
        break
```
